#### Correjimos los tipos de datos
los tipos de datos a cambiar son:
- identificador_paa: str
- codigo_entidad: str
- codigo_municipio: str
- ppto_total: float
- limite_menor_cuantia: float
- limite_minima_cuantia: float
- fecha_ultima_actualizacion: datetime
- fecha_cargue: timestamp

```python
df['identificador_paa']=df['identificador_paa'].astype(str)
df['codigo_entidad']=df['codigo_entidad'].astype(str)
df['identificador_paa']=df['identificador_paa'].astype(str)
df['codigo_municipio']=df['codigo_municipio'].astype(str)
df['ppto_total']= df['ppto_total'].apply(lambda x: x.replace('$','') and x.replace('.','') if isinstance(x, str) else x )
df['limite_menor_cuantia']= df['limite_menor_cuantia'].apply(lambda x: x.replace('$','') and x.replace('.','')  if isinstance(x, str) else x)
df['limite_minima_cuantia']= df['limite_minima_cuantia'].apply(lambda x: x.replace('$','') and x.replace('.','') if isinstance(x, str) else x)
```

```python
def safe_convert_to_float(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return None  # O un valor predeterminado de tu elección
```

```python
df['ppto_total'] = df['ppto_total'].apply(safe_convert_to_float)
df['limite_menor_cuantia']= df['limite_menor_cuantia'].apply(safe_convert_to_float)
df['limite_minima_cuantia']= df['limite_minima_cuantia'].apply(safe_convert_to_float)
```

```python
def safe_convert_to_date(value):
    try:
        return pd.to_datetime(value,format='%m/%d/%y')
    except:
        try:
            return pd.to_datetime(value,format='%d%m%y')
        except:
            try:
                return pd.to_datetime(value,format='%d de %B de %Y')
            except (ValueError, TypeError):
                return None  # O un valor predeterminado de tu elección
```

```python
df['fecha_ultima_actualizacion']=df['fecha_ultima_actualizacion'].apply(safe_convert_to_date)
```

```python
def safe_convert_to_timestamp(value):
    try:
        return pd.Timestamp(value)
    except (ValueError, TypeError):
        return None  # O un valor predeterminado de tu elección
```

```python
df['fecha_cargue']=df['fecha_cargue'].apply(safe_convert_to_timestamp)
```

```python
df['nit_entidad']=df['nit_entidad'].apply(lambda x: x.replace('.',''))
df['nit_entidad']=df['nit_entidad'].apply(lambda x: x[:9])
df['nit_entidad']=df['nit_entidad'].apply(safe_convert_to_float)
df['nit_entidad']=df['nit_entidad'].astype(str)
```

#### Gráfica de barras: Departamentos (frecuencia de registros)#### 

```python
department_counts = secop['departamento_paa'].value_counts()

plt.figure(figsize=(10, 6))
plt.rc('font', size=8) 
department_counts.plot(kind='bar', color='skyblue')
plt.title('Frecuencia de registros por departamento')
plt.xlabel('Departamento')
plt.ylabel('Número de registros')
plt.show()
```

#### Gráfica de pastel: Top 10 departamentos con mayor número de registros

```python
(department_counts.head(10).sum()/department_counts.sum())*100
```

Top 5 departamentos con mayor número de registros representan el 62.5% del total de empresas registradas en el SECOP, a continuación se muestra la participacion de cada uno de esos 10 departamentos dentro de ese porcentaje

```python
top_5_departments = department_counts.head(10)
plt.figure(figsize=(8, 8))
top_5_departments.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Top 10 Departamentos por número de registros')
plt.ylabel('')
plt.show()
```

#### Gráfica de barras: Municipios con más empresas (frecuencia de registros)

```python
municipality_counts = secop['municipio_paa'].value_counts()
top_municipality = municipality_counts.head(5)

plt.figure(figsize=(8, 6))
top_municipality.plot(kind='bar', color='lightgreen')
plt.title('Municipio con más empresas')
plt.xlabel('Municipio')
plt.ylabel('Número de empresas')
plt.xticks(rotation=45)
plt.show()
```

#### Gráfica de barras: Año con más empresas

```python
# Gráfica de barras: Año con más empresas
year_counts = secop['anno'].value_counts()
year_counts = year_counts.sort_index()
plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar', color='orange')
plt.title('Año con más empresas')
plt.xlabel('Año')
plt.ylabel('Número de empresas')
plt.xticks(rotation=0)
plt.show()
```

#### Grafica de Barras: Top 5 Entidades con mayor Presupuesto general

```python
top_5_entidades = secop.sort_values(by='valor_presupuesto_general',ascending=False).head(5)

plt.figure(figsize=(10, 6))
bars=plt.bar(top_5_entidades['nombre_entidad'], top_5_entidades['valor_presupuesto_general'], color='salmon')
plt.title('Top 5 entidades con mayor valor de presupuesto general')
plt.xlabel('Entidad')
plt.ylabel('Valor del presupuesto general')
plt.xticks(rotation=90)

# Agregar los valores numéricos en la parte superior de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

# Mostrar la gráfica
plt.show()
```

#### Gráfico de líneas: Tendencia del número de empresas a lo largo del tiempo

```python
# Agrupar por fecha de creación y contar el número de empresas
secop['fecha_de_creacion'] = pd.to_datetime(secop['fecha_de_creacion'])
trend = secop.groupby(secop['fecha_de_creacion'].dt.to_period('M')).size()

# Gráfico de líneas
plt.figure(figsize=(10, 6))
trend.plot(kind='line', color='green')
plt.title('Tendencia del número de empresas a lo largo del tiempo')
plt.xlabel('Fecha de creación')
plt.ylabel('Número de empresas')
plt.show()
```
