#### Correjimos los tipos de datos
los tipos de datos a cambiar son:
- identificador_paa: str
- codigo_entidad: str
- codigo_municipio: str
- ppto_total: float
- limite_menor_cuantia: float
- limite_minima_cuantia: float
- fecha_ultima_actualizacion: datetime
- fecha_cargue: timestamp

```python
df['identificador_paa']=df['identificador_paa'].astype(str)
df['codigo_entidad']=df['codigo_entidad'].astype(str)
df['identificador_paa']=df['identificador_paa'].astype(str)
df['codigo_municipio']=df['codigo_municipio'].astype(str)
df['ppto_total']= df['ppto_total'].apply(lambda x: x.replace('$','') if isinstance(x, str) else x )
df['limite_menor_cuantia']= df['limite_menor_cuantia'].apply(lambda x: x.replace('$','')  if isinstance(x, str) else x)
df['limite_minima_cuantia']= df['limite_minima_cuantia'].apply(lambda x: x.replace('$','') if isinstance(x, str) else x)
```

```python
def safe_convert_to_float(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return None  # O un valor predeterminado de tu elección
```

```python
df['ppto_total'] = df['ppto_total'].apply(safe_convert_to_float)
df['limite_menor_cuantia']= df['limite_menor_cuantia'].apply(safe_convert_to_float)
df['limite_minima_cuantia']= df['limite_minima_cuantia'].apply(safe_convert_to_float)
```

```python
def safe_convert_to_date(value):
    try:
        return pd.to_datetime(value,format='%m/%d/%y')
    except:
        try:
            return pd.to_datetime(value,format='%d%m%y')
        except:
            try:
                return pd.to_datetime(value,format='%d de %B de %Y')
            except (ValueError, TypeError):
                return None  # O un valor predeterminado de tu elección
```

```python
df['fecha_ultima_actualizacion']=df['fecha_ultima_actualizacion'].apply(safe_convert_to_date)
```

```python
def safe_convert_to_timestamp(value):
    try:
        return pd.Timestamp(value)
    except (ValueError, TypeError):
        return None  # O un valor predeterminado de tu elección
```

```python
df['fecha_cargue']=df['fecha_cargue'].apply(safe_convert_to_timestamp)
```

```python
df['nit_entidad']=df['nit_entidad'].apply(lambda x: x.replace('.',''))
df['nit_entidad']=df['nit_entidad'].apply(lambda x: x[:9])
df['nit_entidad']=df['nit_entidad'].apply(safe_convert_to_float)
df['nit_entidad']=df['nit_entidad'].astype(str)
```
